```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read data
crime = pd.read_csv('Crime_Incidents_in_2022.csv')

# Convert 'REPORT_DAT' to datetime and extract day of the week
crime['REPORT_DAT_dt'] = pd.to_datetime(crime['REPORT_DAT'], utc=True)
crime['Day_of_Week'] = crime['REPORT_DAT_dt'].dt.day_name()

# Count crimes per day of the week
crime_by_day = crime['Day_of_Week'].value_counts()

# Convert to DataFrame for plotting
class_df = pd.DataFrame({'Day_of_Week': crime_by_day.index, 'count': crime_by_day.values})

# Find day with most and least crimes
most_crime_day = crime_by_day.idxmax()
least_crime_day = crime_by_day.idxmin()

print(crime_by_day.to_string())
print(f"Day with most crimes: {most_crime_day} ({crime_by_day.max()})")
print(f"Day with least crimes: {least_crime_day} ({crime_by_day.min()})")

# Plotting
plt.figure(figsize=(16, 6))
sns.barplot(x='Day_of_Week', y='count', data=class_df, palette='bright')
plt.title('Crime Chart')
plt.xlabel('Day of Week')
plt.ylabel('Number of Crimes')
plt.show()
```
